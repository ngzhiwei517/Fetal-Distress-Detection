# Step 1: Load everything

```python
import joblib
import pandas as pd
#model path
model = joblib.load("/content/xgb_ctg_tuned_model.joblib")
scaler = joblib.load("ctg_scaler.joblib")
FEATURES = joblib.load("ctg_features.joblib")
```

# Step 2: Load FULL CSV

```python
import pandas as pd

df_input = pd.read_csv("/content/CTG.csv") #change to your data path
df_input
```

# Step 3: Drop targets + metadata (internally)

```python
DROP_COLS = [
    "Date", "SegFile", "FileName",
    "CLASS", "NSP"
]

X_input = df_input.drop(columns=[c for c in DROP_COLS if c in df_input.columns])
```

# Step 4: Select EXACT training features

```python
X_input = X_input[FEATURES]
```

# Step 5: Numeric view + scaling

```python
X_input = X_input.apply(pd.to_numeric, errors="coerce")
X_input_scaled = scaler.transform(X_input)
```

# Step 6: Predict

```python
pred_raw = model.predict(X_input_scaled)
pred = pred_raw + 1
df=pd.DataFrame(pred,columns=['NSP'])
df
```

# Step 7: Attach predictions

```python
df_input["Predicted_NSP"] = pred
df_input.to_csv("CTG_with_predictions.csv", index=False)
```
